```python
A = assemble(a)
rhs = assemble(L)
AwithBC = assemble(a,bcs=bcs)
rhswithBC = assemble(L,bcs=bcs)

# print options
np.set_printoptions(precision=1)
np.set_printoptions(linewidth=400)
np.set_printoptions(suppress=True)

# print utility functions
def print_matrix(A):
  Apetsc = A.petscmat
  row_idxs = range(0, Apetsc.getSize()[0])
  col_idxs = range(0, Apetsc.getSize()[1])
  print(Apetsc.getValues(row_idxs, col_idxs))
  print() # additional empty line
def print_rhs(b):
  print(b.vector().array())
  print() # additional empty line

print_matrix(A)
print_rhs(rhs)

print_matrix(AwithBC) # Dirichlet rows and columns replaced by identity
print_rhs(rhswithBC)  # Dirichlet entries replaced by 0
```
